```python
df_country = pd.read_excel(r"C:\Users\areeb\Downloads\DataSets\Zomatodataset\Country-Code.xlsx")
df_country.head()
```

```python
final_df = pd.merge(df,df_country,on = "Country Code", how = 'left' )
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
country_names = final_df["Country"].value_counts().index
```

```python
country_values = final_df['Country'].value_counts().values
```

```python
plt.pie(country_values[:3], 
        labels = country_names[:3], 
        autopct="%1.2f%%" ,
        wedgeprops = {"linewidth":1, "edgecolor":"black"})

plt.show()
```

Observation : Maximum Transaction Records are from India

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

# observation
1. 4.5 to 4.9 --> excellent
2. 4.0 to 4.4 --> very good
3. 3.5 to 3.9 --> good
4. 2.5 to 3.4 --> average
5. 1.8 to 2.4 --> poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)

sns.barplot(x="Aggregate rating", 
            y ='Rating Count', 
            data = ratings)
plt.show()
```

```python
sns.barplot(x="Aggregate rating", 
            y ='Rating Count', 
            data = ratings,
            hue = 'Rating color', 
            palette = ['blue','red','orange','yellow','green','green'])

plt.show()
```

observation:

1. not rated count is very high
2. maximun number of ratings is between 2.5 to 3.4

```python
#count plot
sns.countplot(x='Rating color', 
              data = ratings , 
              palette =  ['blue','red','orange','yellow','
              green','green'] ,
              edgecolor = "black",
              linewidth = 2)

plt.show()
```
