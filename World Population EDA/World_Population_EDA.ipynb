```python
top_5 = df.sort_values(by = "2022 Population", ascending = False).head()
```

```python
plt.figure(figsize = (20,8))
           
ax = sns.barplot( x = top_5['Country'] ,
                  y = top_5['2022 Population'])
plt.title("Top 5 Populated Country in 2022")

for i in ax.containers:
    ax.bar_label(i)
```

```python
plt.figure(figsize = (20,8))

ax = sns.barplot( x = top_5['Country'] ,
                  y = top_5['2020 Population'])
plt.title("Top 5 Populated Country in 2020")

for i in ax.containers:
    ax.bar_label(i)
```

```python
plt.figure(figsize = (20,8))

ax = sns.barplot( x = top_5['Country'] ,
                  y = top_5['2015 Population'])
plt.title("Top 5 Populated Country in 2015")

for i in ax.containers:
    ax.bar_label(i)
```
