```python
plt.figure(figsize = (7,7))

plt.bar(
        list(fifa["nationality"].value_counts().keys()[0:5]),
        list(fifa["nationality"].value_counts()[0:5]),
        )

plt.xlabel("Countries")
plt.ylabel("Players")

plt.title("Top 5 freuquency of players Nationality")
plt.show()
```

```python
player_salary = fifa[["short_name","wage_eur"]]
```

```python
player_salary.head()
```

```python
player_salary = player_salary.sort_values(by = ["wage_eur"], ascending = False)
```

```python
player_salary.head()
```

```python
plt.figure(figsize = (8,5))

plt.bar(
        list(player_salary["short_name"])[0:5],
        list(player_salary["wage_eur"])[0:5],
        color = ["red","g","b","y","purple"],
        linewidth = 1,
        edgecolor = "black",
        )

plt.xlabel("Players", size = 20)
plt.ylabel("Salary", size = 20)

plt.title("Top 5Players with highest salary")
plt.show()
```
