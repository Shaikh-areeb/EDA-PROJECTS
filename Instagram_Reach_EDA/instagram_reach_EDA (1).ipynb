```python
#  let’s have a look at all the metrics like Likes, Saves, and Follows from each post over time

plt.figure(figsize = (15,5))
sns.scatterplot(data , x = data.index , y = data["Likes"] , label = "Likes")
sns.scatterplot(data , x = data.index , y = data["Saves"] , label = "Saves")
sns.scatterplot(data , x = data.index , y = data["Follows"], label = "Follows")

plt.legend()
```

```python
# let’s have a look at the distribution of reach from different sources

reach_sources = ['From Home' , 'From Hashtags', 'From Explore', 'From Other']

reach_counts = [data[source].sum() for source in reach_sources]

reach_counts
```

```python
plt.figure(figsize = (15,5))
plt.pie(x = reach_counts , labels = reach_sources, autopct = "%1.2f%%")
plt.legend()
plt.title("Reach From differnt Sources")
plt.show()
```

```python
#  let’s have a look at the distribution of engagement sources

engagements = ["Saves", 'Comments', 'Shares','Likes']

engagement_counts = [data[i].sum() for i in engagements]

engagement_counts
```

```python
plt.figure(figsize = (15,5))
plt.pie(x =  engagement_counts , labels = engagements , autopct = "%1.2f%%")
plt.title('Engagement Sources')
plt.legend()
plt.show()
```

```python
# let’s have a look at the relationship between the number of profile visits and follows

plt.figure(figsize = (15,5))
sns.scatterplot(data,
                 x = "Profile Visits", 
                 y = 'Follows')
plt.title("Profile Visits Vs Follows")
plt.show()
```

```python
corr_matrix = data.corr(numeric_only = True)

plt.figure(figsize = (15,5))
sns.heatmap(corr_matrix )
plt.title("Correlation_Matrix")
plt.show()
```

```python
# let’s explore the hashtags column in detail. Each post contains different combinations of hashtags, which impacts reach on Instagram. So let’s have a look at the distribution of hashtags to see which hashtag is used the most in all the posts

# list to store all hashtags
all_hashtags = []

# Iterate through each row in the 'Hashtags' column
for row in data['Hashtags']:
    hashtags = str(row).split()
    hashtags = [tag.strip() for tag in hashtags]
    all_hashtags.extend(hashtags)

# pandas DataFrame to store the hashtag distribution
hashtag_distribution = pd.Series(all_hashtags).value_counts().reset_index()
hashtag_distribution.columns = ['Hashtag', 'Count']
```

```python
plt.figure(figsize = (18,7))
sns.barplot(hashtag_distribution, x='Hashtag', y='Count')

plt.xticks(rotation = 90)
plt.title('Distribution of Hashtags')
plt.show()
```

```python
# let’s have a look at the distribution of likes and impressions received from the presence of each hashtag on the post

# Create a dictionary to store the likes and impressions for each hashtag
hashtag_likes = {}
hashtag_impressions = {}

# Iterate through each row in the dataset
for index, row in data.iterrows():
    hashtags = str(row['Hashtags']).split()
    for hashtag in hashtags:
        hashtag = hashtag.strip()
        if hashtag not in hashtag_likes:
            hashtag_likes[hashtag] = 0
            hashtag_impressions[hashtag] = 0
        hashtag_likes[hashtag] += row['Likes']
        hashtag_impressions[hashtag] += row['Impressions']

# Create a DataFrame for likes distribution
likes_distribution = pd.DataFrame(list(hashtag_likes.items()), columns=['Hashtag', 'Likes'])

# Create a DataFrame for impressions distribution
impressions_distribution = pd.DataFrame(list(hashtag_impressions.items()), columns=['Hashtag', 'Impressions'])
```

```python
plt.figure(figsize = (15,5))

plt.figure(figsize = (15,5))

sns.barplot(likes_distribution, x='Hashtag', y='Likes')

plt.xticks(rotation = 90);

plt.title('Likes Distribution for Each Hashtag')
plt.show()


plt.figure(figsize = (15,5))

sns.barplot(impressions_distribution, x='Hashtag', y='Impressions')

plt.xticks(rotation = 90);
plt.title('Impressions Distribution for Each Hashtag')
plt.show()
```
